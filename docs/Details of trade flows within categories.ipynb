This notebook gives a template for exploring the data in more detail.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../build/checking_table.csv')
df['commodity'] = df['cmdCode'].astype(str) + ' (' + df['cmdDescE'] + ')'
#df.columns
```

Fore example, this is what goes into sector code "4" imports:

```python
xx = df.query('sector_code == "4" and direction == "Import"').set_index(['year', 'commodity'])['mass_iron'].unstack()
xx.plot(style='.-');
```

And this is the same for sector code "5":

```python
xx = df.query('sector_code == "5" and direction == "Import"').set_index(['year', 'commodity'])['mass_iron'].unstack()
xx.iloc[:, :].plot(style='.-');
plt.legend(bbox_to_anchor=(1, 1));
```
